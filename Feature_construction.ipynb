```python
import numpy as np
import pandas as pd

from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression

import seaborn as sns
```

```python
df=pd.read_csv('train.csv')[['Age','Pclass','SibSp','Parch','Survived']]
```

```python
df.head()
```

```python
df.dropna(inplace=True)
```

```python
df.head()
```

```python
x= df.iloc[:,0:4]
y= df.iloc[:,-1]
```

```python
x.head()
```

```python
y.head()
```

```python
np.mean(cross_val_score(LogisticRegression(),x,y,scoring='accuracy',cv=50))
```

```python
x['Family_size']=x['Parch']+x['SibSp'] + 1
```

```python
x.head()
```

```python
def myfunc(num):
  if num==1:
    #alone
    return 0
  elif num>1 and num<=4:
      #small family
      return 1
  else:
      #large family
      return 2
```

```python
myfunc(4)
```

```python
x['Family_type']=x['Family_size'].apply(myfunc)
```

```python
x.head()
```

```python
x.drop(columns=['SibSp','Parch','Family_size'],inplace=True)
```

```python
x.head()
```

```python
np.mean(cross_val_score(LogisticRegression(),x,y,scoring='accuracy',cv=50))
```
